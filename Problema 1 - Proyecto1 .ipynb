Instrucciones:

Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%


Importando librerias

```python
import pandas as pd
import numpy as np
```

Ventas Ultimos 5 dias

```python
np.random.seed(123456)
data_frame = pd.DataFrame(
    data=np.random.randint(100,1000,(5,5)), #Numeros del 100 al 1000(# de productos vendidos), 5 col y 5 filas
    index=pd.date_range("20220101", periods=5), #Filas con 5 fechas a partir del 2022-01-01
    columns = ["Computadora", "Celular", "Laptop", "TV", "Tablet"]
)
print(data_frame)
```

Creamos segundo Dataframe con los precios de los productos y la suma de los productos vendidos por categorias

```python
data_frame2 = pd.DataFrame()

data_frame2.index = ["Computadora", "Celular", "Laptop", "TV", "Tablet"]
data_frame2["Precio Unitario"] = [20000, 30000, 17000, 18000, 15000]
data_frame2["Unidades"] = [sum(data_frame["Computadora"]), sum(data_frame["Celular"]) ,sum(data_frame["Laptop"]),
            sum(data_frame["TV"]),sum(data_frame["Tablet"])]

print(data_frame2)
```

Creamos la columna Ganancias que es la multiplicacion entre el precio unitario y las unidades

```python
Ganancias = [] #Se crea una lista vacia
for i in range (len(data_frame2)): #i=0 , len=4 
    Ganancias.append(data_frame2["Precio Unitario"][i] * data_frame2["Unidades"][i]) 

data_frame2["Ganancias"] = Ganancias
```

Una segunda forma mas practica de obtener la Ganancias es usando la funcion "multiply"

```python
#data_frame2["Ganancias"] = np.multiply(data_frame2["Precio Unitario"],data_frame2["Unidades"])
```

Creamos la columna Ganancias Desc que es la multiplicacion del precio unitario por unidades y el descuento establecido del 5%

```python
Ganancias_Desc = [] #Se crea una lista vacia
for i in range (len(data_frame2)):
    Ganancias_Desc.append(data_frame2["Precio Unitario"][i] * data_frame2["Unidades"][i] * 0.95)

data_frame2["Ganancias Desc"] = Ganancias_Desc


print(data_frame2)
```
