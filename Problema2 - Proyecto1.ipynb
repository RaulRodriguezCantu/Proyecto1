Instrucciones:

Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa y devuelva una serie con el ingreso mínimo, máximo y media. El resultado final debe ser una serie con los estadísticos.

Llamamos a nuestras librerias

```python
import numpy as np
import pandas as pd
```

Ingresos Mensuales Clientes - 
Creamos un dataframe que se alimenta de un diccionario

```python
Ingresos_Mensuales = pd.DataFrame(data = {
                     'Cliente1': np.random.randint(5000,20000, 12),    #Se alimenta de un diccionario
                     'Cliente2': np.random.randint(5000,20000, 12),
                     'Cliente3': np.random.randint(5000,20000, 12),
                     'Cliente4': np.random.randint(5000,20000, 12),
                     'Cliente5': np.random.randint(5000,20000, 12),
                     },
                     index = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"])

print(Ingresos_Mensuales)
```

Media de clientes

```python
print(Ingresos_Mensuales.mean().round(6))
print(type(Ingresos_Mensuales.mean().round(6))) #Comprobamos que estamos devolviendo una serie
```

Ingreso maximo de cada cliente

```python
print(Ingresos_Mensuales.max().round(6))
print(type(Ingresos_Mensuales.max().round(6))) #Comprobamos que estamos devolviendo una serie
```

Ingreso minimo de cada cliente

```python
print(Ingresos_Mensuales.min().round(6))
print(type(Ingresos_Mensuales.min().round(6))) #Comprobamos que estamos devolviendo una serie
```
